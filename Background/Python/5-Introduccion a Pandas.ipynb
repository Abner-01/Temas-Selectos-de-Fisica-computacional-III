1.- ¿Por qué Pandas?

<font size=4>

Pandas es un herramienta para manipulación y análisis de datos de código abierto rápida, poderosa, flexible y fácil de utilizar, construida sobre el lenguaje de programación Python.\
Es popular en la ciencia de datos y la inteligencia artificial debido a las ventajas y facilidades que proporciona para cargar, preparar, manipular, modelar y analizar datos.

<font size=4>
Algunas características que hacen útil a Pandas son:

<font size=4>
    
1.- Estructuras de datos similares a matrices ($DataFrame$).\
2.- DataFrames similares a los de otros lenguajes.\
3.- Básicamente, los DataFrames son equivalentes a una tabla.\
4.- Los datos en las columnas deben ser del mismo tipo, pero los datos en renglones pueden variar.\
5.- Altamente flexible para tomar datos de múltiples fuentes.

<font size=4> Leemos el DataFramecon pandas:

```python
df = pd.read_csv('data/train.csv')
df  #Mostramos el DataFrame
```

<font size=4> También podemos renombrar una columna en particular utilzando el atributo **.rename()**

```python
df.rename(columns = {"number_of_elements":"numero_de_elementos"},inplace = True)  #Cambiamos el nombre de la columna "C" por "Z"
df
```

<font size=4> También podemos acceder a información de las columnas del dataframe mediante el atributo **df.column_name** o mediante la instrucción **df['column_name']**

```python
#Utilizamos cada uno de los comandos:

df.numero_de_elementos
```

```python
df["numero_de_elementos"]
```

<font size=4> Podemos extraer información de un renglón específico del $DataFrame$  (index) mediante **series** con el atributo **.loc[]**:

```python
df.loc[524]  #Utilizamos el atributo .loc()
```

<font size=4> Se puede acceder a información en particular utilizando los métodos previos, si por ejemplo queremos obtener la temperatura crítica del superconductor 524:

```python
df['critical_temp'][524]
```

<font size=4> En pandas se utilizan también los **operadores de acceso** como el atributo **.iloc[]** que basa su operación en la enumeración de renglones y columnas de Python

<font size=4> En el que por ejemplo de la instrucción **df.iloc[i]** se obtiene la información del renglón **i+1**

```python
df.iloc[524]
```

```python
#Si queremos obtener el numero de elementos (columna 0) de todos los superconductores:
df.iloc[:, 0]
```

<font size=4> El atributo **.loc[]** basa su operación en el índice de los datos, no en su posición o enumeración de Python:

```python
#Si queremos obtener el numero de elementos en el superconductor 524:
df.loc[524, 'numero_de_elementos']
```

```python
#Si quieremos obtener el numero de elementos de los superconductores 524 y 342 por ejemplo:
df.loc[[524, 342], 'numero_de_elementos']
```

```python
#Si quieremos obtener el numero de elemento (columna 0) y temperatura crítica (columna 85) de cada uno de los superconductores:
df.loc[:, ['numero_de_elementos', "critical_temp"]]
```

<font size=4> Pandas también puede utilizar **operadores lógicos** como **==**, **<** y **>**  de la siguiente manera:

```python
#SI queremos obtener los superconductores con temperatura crítica menor a 85 K
df.critical_temp < 85 
```

```python
#Si queremos obtener los superconductores con temperatura crítica igual a 85:
df.loc[df.critical_temp == 85]
```

```python
#SI queremos los superconductores con temperatura crítica de 83 o de 85 K
df.loc[df.critical_temp.isin([85, 83])]
```

<font size=4> Si queremos identificar todas las posiciones en el dataframe que tienen un valor vacío o **"NaN"**

```python
df.loc[df.critical_temp.isnull()].shape
#df.loc[df.critical_temp.notnull()].shape
```

<font size=4> Pandas también puede calcular promedios, desviación estándar, etc. y otras funciones esenciales, por ejemplo:

<font size=4> El atributo **.describe()** nos da la siguiente información de cada una de las columnas del dataframe:

```python
df.describe()
```

```python
#Si queremos sólo la información de la temperatura crítica:
df.critical_temp.describe()
```

<font size=4> El atributo **.mean()** nos regresa el promedio:

```python
#df.critical_temp.mean()
df.mean()
```
